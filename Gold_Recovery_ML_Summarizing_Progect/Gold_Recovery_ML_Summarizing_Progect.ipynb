### Проверка правильности рассчета эффективности обогащения  

```python
# Посмотрим на данные признака 'rougher.output.recovery' в обучающей выборке
df_train['rougher.output.recovery']
```

```python
# Расситаем эффективность обогащения по формуле:
rougher_output_recovery_check = ((df_train['rougher.output.concentrate_au'] 
 * (df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au']))/
 (df_train['rougher.input.feed_au']
 * (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au']))*100)
```

```python
# Посмотрим на получившиеся данные
rougher_output_recovery_check
```

```python
# Расситаем MAE между нашими расчётами и значением признака в тренировочной выборке  
MAE = (sum(abs(rougher_output_recovery_check - df_train['rougher.output.recovery']))
       /len(df_train['rougher.output.recovery']))
```

```python
MAE
```

```python
# Напишем фукцию для подсчета sMAPE
def smape(t, p):
    return 1/len(t)*np.sum(2*np.abs(t-p)/(np.abs(t)+np.abs(p))*100)
```

### Случайный лес

```python
# Инициализируем модель случайного леса для финального концентрата, 
# настроим гиперпараметры с помощью optuna и проверим качество модели на кросс-валидации с помощью cross_val_score

def objective(trial):
    max_depth = trial.suggest_int('max_depth', 1, 20, 2)
    min_samples_split = trial.suggest_int('min_samples_split', 1, 20, 2)
    n_estimators = trial.suggest_int('n_estimators', 10, 50, 10)
    model = RandomForestRegressor(max_depth =  max_depth, 
                                  min_samples_split = min_samples_split, 
                                  n_estimators = n_estimators, 
                                  random_state = 12345)
    sc = cross_val_score(model, features_train, target_train_final, n_jobs = -1, cv = 5, scoring = score)
    return - sc.mean()

study = optuna.create_study(direction ='minimize')
study.optimize(objective, n_trials = 25, show_progress_bar = True)

max_depth_fin = study.best_trial.params.get('max_depth')
min_samples_split_fin = study.best_trial.params.get('min_samples_split')
n_estimators_fin = study.best_trial.params.get('n_estimators')
forest_fin_best_score = study.best_value

forest_fin = RandomForestRegressor(max_depth =  max_depth_fin, 
                                   min_samples_split = min_samples_split_fin, 
                                   n_estimators = n_estimators_fin)
```

Качество моделей мы оценивали метрикой `sMAPE` (Симметричное среднее абсолютное процентное отклонение). Мы написали функцию для ее расчёта по формуле:
