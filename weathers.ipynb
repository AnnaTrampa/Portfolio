```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import re
```

```python
train_df = pd.read_csv('train_dataset.csv')
train_df.head()
```

```python
weathers = train_df['weather_pred'].dropna().unique().tolist()
weathers
```

```python
weathers_letters = [re.sub('[^а-яА-Я ]+', '', s) for s in weathers]
weathers_letters
```

```python
len(set(weathers_letters))
```

```python
[x.strip() for x in set(weathers_letters)]
```

```python
bad_weathers = {'пасм': 'overcast', 'дожд': 'rain', 'снег': 'snow', 'ливень': 'heavy_rain', 'гроз': 'thunder', 'шторм': 'storm'}
for weather_type in bad_weathers.keys():
    train_df[bad_weathers[weather_type]] = np.where(train_df['weather_pred'].str.contains(weather_type), 1, 0)

train_df['cloudy'] = np.where((train_df['weather_pred'].str.contains(' обл')) | (train_df['weather_pred'].str.startswith('обл')), 1, 0)
```
